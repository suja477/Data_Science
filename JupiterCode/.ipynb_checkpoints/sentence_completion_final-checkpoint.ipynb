```python
data=A.read()
```

```python
import re
```

```python
data1=re.sub("\n"," ",data)
```

```python
data2=data1.upper()
```


# convert data in x and y format

```python
tokenized= data2.split(" ")
```

```python
tokenized
```

```python
len(tokenized)
```

```python
for i in range(3,234,1):
    w1=tokenized[i-3]
    w2=tokenized[i-2]
    w3=tokenized[i-1]
    w4=tokenized[i]
    print(w1,w2,w3,"-->",w4)
```

```python
inp = []
op = []
for i in range(2,234,1):
    w1 = tokenized[i-2]
    w2 = tokenized[i-1]
    w3 = tokenized[i]
    inp.append(w1+" "+w2)
    op.append(w3)
```

```python
import pandas as pd
Q = pd.DataFrame([inp,op]).T
Q.columns=["Input","Output"]
```

```python
Q
```
