# Determining circumference of the circle using class and object


# C = 2 \pi r
C	=	circumference
\pi	=	the constant pi
r	=	radius of the circle


```python
class circle:
    
    pi = 5.90
    
    def __init__(self,radius=1):
        self.radius = radius
        
        
    def perimeter(self):
        print('circumference of the circle is : ',2*self.pi*self.radius)
```

```python
c1 = circle()
c1.perimeter()
```

```python
c2 = circle(5)
c2.perimeter()
```
