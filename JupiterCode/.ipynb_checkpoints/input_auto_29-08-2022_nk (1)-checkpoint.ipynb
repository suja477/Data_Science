```python
class auto_ppt:
    def __init__(self):
        pass
    def data_file(self, filename):
        if filename.split('.')[-1] == 'csv':
        # Assume that the user uploaded a CSV file
            flag = 0
            while(flag == 0):
                try:
                    df = pd.read_csv(filename)
                    flag = 1

                except:
                    print('Error reading the file')

                    try:
                        df = pd.read_csv(filename, encoding='utf-sig-8')
                        flag = 1

                    except:
                        print('Error reading the file')

                        try:
                            df = pd.read_csv(filename, encoding='windows-1252')
                            flag = 1

                        except:
                            print('Error reading the file')

                            try:
                                df = pd.read_csv(filename, encoding='utf-8')
                                flag = 1

                            except:
                                print('Error reading the file')

        elif filename.split('.')[-1] == 'xls':
        # Assume that the user uploaded an excel file
            flag = 0
            while(flag == 0):
                try:
                    df = pd.read_csv(filename)
                    flag = 1

                except:
                    print('Error reading the file')

                    try:
                        df = pd.read_csv(filename, encoding='utf-sig-8')
                        flag = 1

                    except:
                        print('Error reading the file')

                        try:
                            df = pd.read_csv(filename, encoding='windows-1252')
                            flag = 1

                        except:
                            print('Error reading the file')

                            try:
                                df = pd.read_csv(filename, encoding='utf-8')
                                flag = 1

                            except:
                                print('Error reading the file')

        elif filename.split('.')[-1] == 'xlsx':
                # Assume that the user uploaded an excel file
            flag = 0
            while(flag == 0):
                try:
                    df = pd.read_csv(filename)
                    flag = 1

                except:
                    print('Error reading the file')

                    try:
                        df = pd.read_csv(filename, encoding='utf-sig-8')
                        flag = 1

                    except:
                        print('Error reading the file')

                        try:
                            df = pd.read_csv(filename, encoding='windows-1252')
                            flag = 1

                        except:
                            print('Error reading the file')

                            try:
                                df = pd.read_csv(filename, encoding='utf-8')
                                flag = 1

                            except:
                                print('Error reading the file')
        return df

    def user_input(self,data):
    #     print(data_1.columns)
        data_1=data.copy()
        data_1.columns=data_1.columns.str.replace('_',' ')
        target_list = []
        for i, j in zip(data_1.columns, data_1.dtypes):
    #         print(i, ':\t', j)
            if j == 'int64' or j == 'float64' or j == 'object' :
                a = 2
                if data_1[i].nunique() >=2:
                    data_1[i] = data_1[i].astype(object)
                    target_list.append(i)
                else:
                    pass

            else:
                pass

        data2 = data_1[target_list]
        data2['None'] = np.nan
        #printing dataframe of target variable
        print("Target Dataframe:")
        for i,col in enumerate(data2.columns):
            print(i,col)
        flag=0
        while (flag==0):
            try:
                y=int(input('Enter the target column: '))
                
                if data2.iloc[:,y].name == 'None':
                    flag=1
                    Y = "None"
                    
                else:
                    flag=1
                    Y=data2.iloc[:,y]
#                     print(type(Y))
            except:
                print('Enter valid input')

        #Printing dataframe of independent variable
        print("Independent features dataframe:")
        for i,col in enumerate(data_1.columns):
            print (i,col)
        col_lst=[]
        while (flag==1):
            try:
                x=int(input('Enter the independant column: '))
                if x == y:
                    print('Select column other than Target')
                    flag=1
                z = input('Do you want another input(y/n)?: ')
                col_lst.append(data_1.iloc[:,x].name)
                if z == 'Y' or z == 'y' :
                    flag=1
                else:
                    flag=2

            except:
                print('Enter valid input')
        X=data_1.loc[:,col_lst]
        
        if str(type(Y)) == "<class 'pandas.core.series.Series'>":
            print('\nTarget=',Y.name,'\nIndependant Variable=',col_lst)
            return Y.name, col_lst
        else:
            print('\nTarget=',Y,'\nIndependant Variable=',col_lst)
            return Y, col_lst
#--------------------------------------------------------------------------------------------------

    def positions(self, independant1):
        if len(independant1) == 4:
            pos1 = [1, 0.5, 3.5, 1.7]
            pos2 = [4.5, 0.5, 3.5, 1.7]
            pos3 = [1, 2.7, 3.5, 1.7]
            pos4 = [4.5, 2.7, 3.5, 1.7]
            pos = [pos1, pos2, pos3, pos4]

            pos1t = [1, 0.5, 3.5, 1.7]
            pos2t = [4.5, 0.5, 3.5, 1.7]
            pos3t = [1, 2.7, 3.5, 1.7]
            pos4t = [4.5, 2.7, 3.5, 1.7]
            post = [pos1t, pos2t, pos3t, pos4t]
                    
        elif len(independant1) == 3:
            pos1 = [1.4, 0.6, 5.5, 3.6]
            pos2 = [8.3, 0.6, 5.5, 3.6]
            pos3 = [5.5, 4.8, 5.5, 3.6]
            pos = [pos1, pos2, pos3]

            pos1t = [1.4, 3.8, 5.5, 0.4]
            pos2t = [8.3, 3.8, 5.5, 0.4]
            pos3t = [5.5, 7.9, 5.5, 0.4]
            post = [pos1t, pos2t, pos3t]
                    
        elif len(independant1) == 2:
            pos1 = [0.5, 1.2, 5.5, 4]
            pos2 = [6.7, 1.2, 5.5, 4]
            pos = [pos1, pos2]

            pos1t = [0.7, 4.9, 5, 0.5]
            pos2t = [6.9, 4.9, 5, 0.5]
            post = [pos1t, pos2t]

        else:
            pos1 = [3.5, 1.2, 6.5, 4]
            pos = [pos1]

            pos1t = [3.5, 4.9, 6.5, 0.5]
            post = [pos1t]
        
        return pos, post
    
    def add_text(self, slide, text, left, top, width, height, font_size):
        txBox = slide.shapes.add_textbox(left, top, width, height)
        tf = txBox.text_frame

        p = tf.add_paragraph()
        p.alignment = PP_ALIGN.CENTER
        p.text = text
        p.font.size = Pt(font_size)
        
    
    def create_ppt(self, data1, TARGET, independant):
        data1.columns=data1.columns.str.replace('_',' ')
        summary=[]  
        # Creating ppt object
        path = 'hansa_template.pptx'
        prs = Presentation(path)
#         prs.slide_width = Inches(16)
#         prs.slide_height = Inches(9)
        slide = {}
        
        title_slide_layout =prs.slide_layouts[0]
        slide[0] = prs.slides.add_slide(title_slide_layout)
#         self.add_text(slide[0], 'AUTOMATED PPT', Inches(1), Inches(3), Inches(14), Inches(2), 70)
#         self.add_text(slide[0], TARGET+" ANALYSIS", Inches(1), Inches(4.5), Inches(14), Inches(1), 40)
        
        title = slide[0].shapes.title
        subtitle = slide[0].placeholders[1]
        title.text = "AUTOMATED PPT"
        title.text_frame.paragraphs[0].font.name = "Calibri"
        subtitle.text = TARGET+" ANALYSIS"
        subtitle.text_frame.paragraphs[0].font.name = "Calibri"
        today_date = slide[0].placeholders[17]
        today = date.today()
        today_date.text = 'Date: ' + str(today)
        today_date.text_frame.paragraphs[0].font.name = "Calibri"
#         for shape in slide[0].placeholders:
#             print("_____#####*****"*5)
#             print('%d %s' % (shape.placeholder_format.idx, shape.name))
            
        # dividing idependant feature into list of list containing 4 value
        list1=[]
        list2=[]
        for i,j in enumerate(independant):
            if len(list2)==2:
                list2=[]
                list2.append(j)
            else:
                list2.append(j)
                if len(list2)==2:
                    list1.append(list2)
                    list2=[]
         
        if len(list2)>0:
            list1.append(list2)
            
        #datetime identification
        date_time_col=[]
        for col_123 in independant:
            try:
                if data1[col_123].apply(str).str.contains('-')[0]:
                    data1[col_123] =  pd.to_datetime(data1[col_123])
                    date_time_col.append(col_123)
            except:
                print(col_123,'NOT a date time column')
    
        # first loop for no of slide
        for j,k in enumerate(list1):
            first_slide = prs.slide_masters[0].slide_layouts[2]
            slide[j+1] = prs.slides.add_slide(first_slide)
            independant1=k
            
            
            # second loop for no of graph in a slide
            for i, col in enumerate(independant1):
                # checking for numeric and object data type
                if data1[col].dtype!= 'object':
                    if data1[col].nunique() <= 3:
                        data1['temp'] = data1[col].apply(str)
                        
                    else:
                        print('binning',col)
                        data1['temp']=pd.qcut(data1[col],5,retbins=0)
                        data1['temp']=data1['temp'].astype(object)
                        data1['temp']=data1['temp'].apply(str)
                else:
#                     print(col,'is not numeric')
                    data1['temp']=data1[col]
                
                #contains all the categories
                all_c = list(data1['temp'].value_counts().reset_index()['index'])
#                 print(len(all_c))
                a = list(data1['temp'].value_counts().reset_index()['index'][:5])
#                 print(a)
                data12 = data1[(data1['temp'].isin(a))]
                
                # define chart data ---------------------
                chart_data = CategoryChartData()
                
#                 print('\n\n\n\n\n\n',data12[TARGET].nunique())
                
                
                # for datetime independat var
                print(TARGET != 'None',data12[TARGET].dtype!='object',k[0] in date_time_col,k)
                if TARGET != 'None' and data12[TARGET].dtype!='object' and k[0] in date_time_col:
                    print('hererererer time series')
                    data12["month"]=data12[k[0]].dt.strftime('%m')
                    data12[['month']+[TARGET]].groupby(by='month').sum().reset_index().month.values
                    chart_data = ChartData()
                    chart_data.categories = data12[['month']+[TARGET]].groupby(by='month').sum().reset_index()\
                    .month.values

                    print(data12[['month']+[TARGET]].groupby(by='month').sum().reset_index()\
                    .month.values)

                    chart_data.add_series(str(TARGET),data12[['month']+[TARGET]]\
                    .groupby(by='month').sum().reset_index()[TARGET].values)

                    print(chart_data.add_series(str(TARGET),data12[['month']+[TARGET]]\
                    .groupby(by='month').sum().reset_index()[TARGET].values))
##                         x, y, cx, cy = Inches(2), Inches(2), Inches(6), Inches(4.5)
#                         chart = slide.shapes.add_chart(XL_CHART_TYPE.LINE, x, y, cx, cy, chart_data).chart

                
                elif TARGET != 'None' and data12[TARGET].dtype!='object' and data12[TARGET].nunique()<3  :
                    # Creating dataframe
                    print('2222222')
                    aa=(((pd.crosstab(data12['temp'],data12[TARGET])).sort_values(by=1,ascending=False))).reset_index()
                    print(aa.iloc[:,0].values)
                    chart_data.categories=aa.iloc[:,0].values


                    for bb,item in enumerate(data1[TARGET].unique()):
                        chart_data.add_series(str(item),(aa.loc[:,item]/data1.loc[data1[TARGET]==item,:].shape[0])*100)
                        
                # for two numeric value
                elif TARGET != 'None' and data12[TARGET].dtype!='object' and data12[TARGET].nunique()>=3:
                    print('333333')
                    chart_data = XyChartData()
                    series_1 = chart_data.add_series('Model 1')
                    for e,z in zip(list(data12[TARGET]),list(data12[col].values)): 
                        series_1.add_data_point(e, z)
                    
                    
                else:
                    print('444444')
                    #when target is not selected
                    vc = data12['temp'].value_counts().reset_index()
                    vc['PERCENT'] = (vc['temp'] / vc['temp'].sum()) * 100
                    chart_data.categories=vc.iloc[:,0].values
                    
                    chart_data.add_series('', vc['PERCENT'].tolist())
                    
                pos, post = self.positions(independant1)  
                
                x1, y1, cx1, cy1 = Inches(pos[i][0]), Inches(pos[i][1]), Inches(pos[i][2]), Inches(pos[i][3])
        #         slide.shapes.add_chart(XL_CHART_TYPE.COLUMN_CLUSTERED, x, y, cx, cy, chart_data)
                
                if TARGET != 'None' and data12[TARGET].dtype!='object' and k[0] in date_time_col:
                    graphic_frame = slide[j+1].shapes.add_chart(XL_CHART_TYPE.LINE, x1, y1, cx1, cy1, chart_data).chart
                elif TARGET != 'None' and data12[TARGET].dtype!='object' and data12[TARGET].nunique()>=3:
                    graphic_frame = slide[j+1].shapes.add_chart(XL_CHART_TYPE.XY_SCATTER, x1, y1, cx1, cy1, chart_data).chart
                
                elif 3 <= len(all_c) < 5 and TARGET == 'None':
                    graphic_frame = slide[j+1].shapes.add_chart(XL_CHART_TYPE.DOUGHNUT,\
                                                              x1, y1, cx1, cy1, chart_data)
                
                elif len(all_c) < 3:
                    graphic_frame = slide[j+1].shapes.add_chart(XL_CHART_TYPE.BAR_CLUSTERED,\
                                                              x1, y1, cx1, cy1, chart_data)
                else:
                    graphic_frame = slide[j+1].shapes.add_chart(XL_CHART_TYPE.COLUMN_CLUSTERED,\
                                                              x1, y1, cx1, cy1, chart_data)
                
                
                x2, y2, cx2, cy2 = Inches(post[i][0]), Inches(post[i][1]), Inches(post[i][2]), Inches(post[i][3])
                txBox = slide[j+1].shapes.add_textbox(x2, y2, cx2, cy2)
                tf = txBox.text_frame
                
                
                ### text is insert here for insights
                try:
                    if TARGET != 'None' and data12[TARGET].dtype!='object' and data12[TARGET].nunique()>=3:
                        print('numeric')
                    elif TARGET != 'None':
        #                 p.text = str(col)+" "+str(aa.iloc[0,0])+" "+str(aa.loc[0,1]/data1.loc[data1[TARGET]==item,:].shape[0])+" "+'percent is least performer \n'
    #                     print(aa)
                        p = tf.add_paragraph()
                        aab=(((pd.crosstab(data12['temp'],data12[TARGET],normalize='columns')).sort_values(by=1,ascending=False))).reset_index()
                        value_rand=np.random.rand()
                        if value_rand > 0.9:
                            p.text = str(col)+" "+str(aab.iloc[0,0])+" with "+str(round(aab.loc[0,1]*100,0))+" "+'percent is highest performer.'

                        elif value_rand > 0.8:
                            p.text = str(col)+" "+str(aab.iloc[0,0])+" with "+str(round(aab.loc[0,1]*100,0))+" "+' % has highest impact.'

                        elif value_rand > 0.7:
                            p.text = str(col)+" "+str(aab.iloc[0,0])+" with "+str(round(aab.loc[0,1]*100,0))+" "+'%  most important criteria.'

                        elif value_rand > 0.6:
                            p.text = str(col)+" "+str(aab.iloc[0,0])+" with "+str(round(aab.loc[0,1]*100,0))+" "+'percent has highest share.'

                        elif value_rand > 0.5:
                            p.text = str(col)+" "+str(aab.iloc[0,0])+" with "+str(round(aab.loc[0,1]*100,0))+" "+'percent is the highest category.'

                        elif value_rand > 0.4:
                            p.text = str(col)+" "+str(aab.iloc[0,0])+" with "+str(round(aab.loc[0,1]*100,0))+" "+'% has shown most conversion.'

                        elif value_rand > 0.3:
                            p.text = str(col)+" "+str(aab.iloc[0,0])+" with "+str(round(aab.loc[0,1]*100,0))+" "+'percent of all category'

                        elif value_rand > 0.2:
                            p.text = str(col)+" "+str(aab.iloc[0,0])+" with "+str(round(aab.loc[0,1]*100,0))+" "+'percent of whole data.'

                        elif value_rand > 0.1:
                            p.text = str(col)+" "+str(aab.iloc[0,0])+" with "+str(round(aab.loc[0,1]*100,0))+" "+'percent is of highestest important.'

                        else:
                            p.text = str(col)+" "+str(aab.iloc[0,0])+" with "+str(round(aab.loc[0,1]*100,0))+" "+'percent is peak performer.'


                        p.font.size = Pt(15)
    #                     tf.autofit_text()
                        tf.word_wrap=True
                        p.font.name = "Calibri"

                        summary.append(p.text)
                except:
                    print('Error in insights')
                try:
                    if  i==0:   
                        title = slide[j+1].shapes.title
                        title.text = p.text
                    chart = graphic_frame.chart
                    chart.font.size = Pt(12)

                    if TARGET != 'None':
                        chart.chart_title.text_frame.text = col + ' vs ' + TARGET
                    else:
                        chart.chart_title.text_frame.text = col

                    if TARGET != 'None':
                        pass

                    else:
                        color_list = ["FFCCE0", "E6CCFF", "00CCCC", "FFFFCC", "00CC00"]
                        for idx, point in enumerate(chart.series[0].points):
                            col_idx = idx % len(color_list)
                            point.format.fill.solid()
                            point.format.fill.fore_color.rgb = RGBColor.from_string(color_list[col_idx])
                    chart.has_legend = True
                    chart.legend.position = XL_LEGEND_POSITION.BOTTOM
                    chart.legend.include_in_layout = False
                    
                    if len(all_c) < 3 or len(all_c) >= 5 or TARGET != 'None':
                        value_axis = chart.value_axis
                        value_axis.has_major_gridlines = False
                        
                    plot = chart.plots[0]
                    plot.has_data_labels = True
                    data_labels = plot.data_labels

                    data_labels.font.size = Pt(10)
                    data_labels.font.color.rgb = RGBColor(0x00, 0x00, 0x00)

                    data_labels.number_format = "0\%"

                    data_labels.position = XL_LABEL_POSITION.OUTSIDE_END
                except:
                    print('error in ppt heading')

                
            
              

                
#                     value_axis.visible = False

#                     category_axis = chart.category_axis
#                     category_axis.has_minor_gridlines = False
                
#                 else:
#                     pass
#                     category_axis.visible = False

              
        
        #############################################################################################################
        summary_slide =prs.slide_layouts[1]
        slide[len(list1)+1] = prs.slides.add_slide(summary_slide)
        title = slide[len(list1)+1].shapes.title
        title.text = 'SUMMARY'
        title.text_frame.paragraphs[0].font.name = "Calibri"
        title.text_frame.paragraphs[0].alignment = PP_ALIGN.CENTER
        
        title.text_frame.paragraphs[0].font.size = Pt(40)
        title.text_frame.paragraphs[0].font.underline = True

        subtitle = slide[len(list1)+1].placeholders[1]
        abcde=""
        for i in summary:
            abcd=i
            abcde=abcde+'\n'+abcd
           
        subtitle.text =abcde    
        subtitle.text_frame.paragraphs[0].font.name = "Calibri"
        ##############################################################################################################
        thank_you_slide =prs.slide_layouts[10]
        slide[len(list1)+2] = prs.slides.add_slide(thank_you_slide)
#         self.add_text(slide[len(list1)+1], 'THANK YOU', Inches(1), Inches(2), Inches(11), Inches(1.5), 70)
        
#         slide[len(list1)+1]= prs.slides.add_slide(title_slide_layout)
        title = slide[len(list1)+2].shapes.title
        subtitle = slide[len(list1)+2].placeholders[0]
        title.text = 'THANK YOU'
        title.text_frame.paragraphs[0].font.name = "Calibri"
#         subtitle.text = '-By SV & NK'
#         print(summary)
        prs.save('output_v1.1.pptx')

        
        print("PPT Created")
        os.startfile("Output_v1.1.pptx")
        print('\nOpening pptx File')
        print(date_time_col)
#         shell_process = subprocess.Popen(['outp\ut_v1.1.pptx'],shell=True) 
#         print(shell_process.pid)
#         time.sleep(15)
        
#         parent = psutil.Process(shell_process.pid)
#         children = parent.children(recursive=True)
# #         print(children)
#         child_pid = children[0].pid
# #         print(child_pid) 
#         try:
#             os.kill(child_pid, signal.SIGTERM)
#         except:
#             print('Presentaion Closed Already')
```
