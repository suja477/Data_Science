```python
import pandas as pd
A=pd.read_csv("C:/Users/Suja4/OneDrive/Documents/Data Science/Dataset/Cars93.csv")
from warnings import filterwarnings 
filterwarnings("ignore")
```

```python
from Wd8pm import replacer
replacer(A)
```

```python
Y = A[["Origin"]]
X = A.drop(labels=["id","Make","Model","Origin"],axis=1)
```

```python
cat = []
con = []
for i in X.columns:
    if(X[i].dtypes == "object"):
        cat.append(i)
    else:
        con.append(i)
```

# Preprocessing

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
Ynew = pd.DataFrame(le.fit_transform(Y.Origin),columns=["Origin"])
```

```python
from sklearn.preprocessing import MinMaxScaler
mm = MinMaxScaler()
X1 = pd.DataFrame(mm.fit_transform(X[con]),columns=con)
X2 = pd.get_dummies(X[cat])
Xnew = X1.join(X2)
```

# Create NN1

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
nn = Sequential()
nn.add(Dense(units=100,input_dim=(len(Xnew.columns)))) # input
nn.add(Dense(100,activation="relu")) # hidden
nn.add(Dense(1,activation="sigmoid")) # output
```

```python
nn.compile(loss="mse")
model = nn.fit(Xnew,Ynew,validation_split=0.2,epochs=30)
```

```python
nn.compile(loss="binary_crossentropy",metrics="accuracy")
model = nn.fit(Xnew,Ynew,validation_split=0.2,epochs=10)
```

```python
tr = model.history['accuracy']
```

```python
ts = model.history['val_accuracy']
```

```python
import matplotlib.pyplot as plt
plt.plot(tr,c="blue")
plt.plot(ts,c="red")
plt.xticks(range(1,12,1))
```

# NN2(50 epochs)

```python
nn.compile(loss="binary_crossentropy",metrics="accuracy")
model = nn.fit(Xnew,Ynew,validation_split=0.2,epochs=50)
tr = model.history['accuracy']
ts = model.history['val_accuracy']
```

```python
import matplotlib.pyplot as plt
plt.plot(tr,c="blue")
plt.plot(ts,c="red")
plt.xticks(range(1,50,1))
```

# Configuring Batch Size¶

```python
nn = Sequential()
nn.add(Dense(units=100,input_dim=(len(Xnew.columns)))) # input
nn.add(Dense(100,activation="relu")) # hidden
nn.add(Dense(1,activation="sigmoid")) # output

nn.compile(loss="binary_crossentropy",metrics="accuracy")
model = nn.fit(Xnew,Ynew,validation_split=0.2,epochs=50,batch_size=30)
tr = model.history['accuracy']
ts = model.history['val_accuracy']
```
