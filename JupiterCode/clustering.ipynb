```python
import pandas as pd
A=pd.read_csv("C:/Users/Suja4/OneDrive/Documents/Data Science/Dataset/train.csv")

from warnings import filterwarnings
filterwarnings("ignore")
```

```python
B = A[["Price","MPG.city"]]
B.head(2)
```

```python
from sklearn.preprocessing import MinMaxScaler
mm = MinMaxScaler()
C = pd.DataFrame(mm.fit_transform(B),columns=[B.columns])
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=3)
model = km.fit(C)
```

```python
model.labels_
```

```python
import matplotlib.pyplot as plt
plt.scatter(B['Price'],B['MPG.city'])
plt.xlabel("Price of Car")
plt.ylabel("Mileage of car")
plt.title("Price vs Mileage scatterplot")
```

```python
Q = pd.DataFrame(mm.inverse_transform(model.cluster_centers_),columns=["X","Y"])
Q
```

```python
import matplotlib.pyplot as plt
plt.scatter(B['Price'],B['MPG.city'],s=10,c=model.labels_)
plt.xlabel("Price of Car")
plt.ylabel("Mileage of car")
plt.title("Price vs Mileage scatterplot")
plt.scatter(Q.X,Q.Y,c="red")
```

```python
K = range(2,20,1)
WCSS = []
for i in K:
    from sklearn.cluster import KMeans
    km = KMeans(n_clusters=i)
    model = km.fit(C)
    WCSS.append(model.inertia_)
```

```python
import matplotlib.pyplot as plt
plt.scatter(K,WCSS,s=10,c="red")
plt.plot(K,WCSS,c="blue")
plt.xlabel("No of clusters")
plt.ylabel("WCSS")
plt.title("Elbow Curve")
plt.xticks(range(2,21,1))
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=6)
model = km.fit(C)
```

```python
B['cluster']=model.labels_
B['Model']=A.Model
B.head(10)
```

```python
m = input("Enter your car Model: ")
cluster_no = list(B[B.Model==m].cluster)[0]
simi = list(B[(B.cluster == cluster_no)&(B.Model != m)].Model)
print("For the given car: ",m)
print("---------------------")
print("Similar cars are: \n",simi)
```
