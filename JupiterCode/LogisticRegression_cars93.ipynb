# Bivariate Analysis

```python
from Wd8pm import ANOVA,chisq
```

```python
A = A.rename(columns = {"Min.Price":"Min_Price","Max.Price":"Max_Price","MPG.city":"MPG_city","MPG.highway":"MPG_highway"
    ,"Rev.per.mile":"Rev_per_mile","Man.trans.avail":"Man_trans_avail","Fuel.tank.capacity":"Fuel_tank_capacity",
                        "Turn.circle":"Turn_circle","Rear.seat.room":"Rear_seat_room","Luggage.room":"Luggage_room"})
X = X.rename(columns = {"Min.Price":"Min_Price","Max.Price":"Max_Price","MPG.city":"MPG_city","MPG.highway":"MPG_highway"
    ,"Rev.per.mile":"Rev_per_mile","Man.trans.avail":"Man_trans_avail","Fuel.tank.capacity":"Fuel_tank_capacity",
                        "Turn.circle":"Turn_circle","Rear.seat.room":"Rear_seat_room","Luggage.room":"Luggage_room"})
```

```python
A.head()
```

```python
for i in X.columns:
    if(X[i].dtypes=="object"):
        pvalue=chisq(A,'Origin',i)
        print(i ,"---Chisq---", pvalue)
    else:
        pvalue=ANOVA(A,'Origin',i)
        print(i ,"--------Anova---", pvalue)
```

```python
imp_cols = []
from Wd8pm import ANOVA,chisq
for i in X.columns:
    if(X[i].dtypes == "object"):
        pval = chisq(A,"Origin",i)
        print("Origin ~",i,"---->",pval)
        if(pval<0.05):
            imp_cols.append(i)
    else:
        pval = ANOVA(A,"Origin",i)
        print("Origin ~",i,"---->",pval)
        if(pval<0.05):
            imp_cols.append(i)
```

```python
imp_cols
```

# Preprocessing

```python
Ximp = X[imp_cols]
```

```python

from Wd8pm import preprocessing
Xnew =preprocessing(Ximp)
```

# Split

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(Xnew,Y,test_size=0.2,random_state=21)
```

# Model 

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
model=lr.fit(xtrain,ytrain)
```

# Prediction

```python
tr_pred= model.predict(xtrain)
ts_pred=model.predict(xtest)
```

# Accuracy

```python
from sklearn.metrics import accuracy_score,confusion_matrix
accuracy_score(ytrain,tr_pred) ,accuracy_score(ytest,ts_pred)
```

```python
from sklearn.metrics import confusion_matrix,recall_score
confusion_matrix(ytrain,tr_pred)
```

```python
confusion_matrix(ytest,ts_pred)
```
