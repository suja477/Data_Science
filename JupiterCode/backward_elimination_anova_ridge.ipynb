# Model

```python

from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(Xnew,Y,random_state=21,test_size=0.2)

from statsmodels.api import OLS,add_constant
xconst = add_constant(xtrain) #this is done to get b0 value(not required in LM)
ols = OLS(ytrain,xconst).fit()
round(ols.rsquared_adj,7),ols.pvalues.sort_values().index[-1]
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
model = lm.fit(xtrain,ytrain)
tr_pred = model.predict(xtrain)
ts_pred = model.predict(xtest)
```

```python
from sklearn.metrics import mean_absolute_error
tr_err = mean_absolute_error(ytrain,tr_pred)
ts_err = mean_absolute_error(ytest,ts_pred)
```

```python
tr_err,ts_err
```

# Ridge

```python
from sklearn.linear_model import Ridge
rr = Ridge(alpha=1.6)
model = rr.fit(xtrain,ytrain)
tr_pred = model.predict(xtrain)
ts_pred = model.predict(xtest)
tr_err = mean_absolute_error(ytrain,tr_pred)
ts_err = mean_absolute_error(ytest,ts_pred)
tr_err,ts_err
```

```python
x = 0.01
alphas = []
for i in range(0,400,1):
    alphas.append(x)
    x = round(x + 0.01,2)
```

```python
tr = []
ts = []
for i in alphas:
    rr = Ridge(alpha=i)
    model = rr.fit(xtrain,ytrain)
    tr_pred = model.predict(xtrain)
    ts_pred = model.predict(xtest)
    tr_err = mean_absolute_error(ytrain,tr_pred)
    ts_err = mean_absolute_error(ytest,ts_pred)
    tr.append(tr_err)
    ts.append(ts_err)
```

```python
import matplotlib.pyplot as plt
plt.plot(tr,c="blue")
plt.plot(ts,c="red")
```

```python
alphas[-1]
```

```python
rr = Ridge(alpha=4)
model = rr.fit(xtrain,ytrain)
tr_pred = model.predict(xtrain)
ts_pred = model.predict(xtest)
tr_err = mean_absolute_error(ytrain,tr_pred)
ts_err = mean_absolute_error(ytest,ts_pred)
tr_err,ts_err
```
